```python
# importing of dataset
import pandas as pd
import numpy as np
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# dropping of duplicate values

df.drop_duplicates(keep = False, inplace = True)
```

```python
#Getting dataset of male category only 

df1 = df[df["Athlete gender"] == 'M']
df1
```

```python
#Filtering distances in miles and kilometres only

df2 = df1.loc[df1['Event distance/length'].str.contains('km') | df1['Event distance/length'].str.contains('mi')]
df2
```

```python
# Replacing of all ',' with '.'

df2['Event distance/length'] = df2['Event distance/length'].str.replace(',', '.')
df2
```

```python
# creating new column for 'Event distance/length in km'

df2['Event distance/length in km'] = df.apply(lambda _: None, axis = 1)
```

```python
# creating new column for 'Event distance/length in mi'

df2['Event distance/length in mi'] = df.apply(lambda _: None, axis = 1)

df2
```

```python
# separating the km and miles unit into their different columns

for i in range(len(df2['Event distance/length'])):
    if 'km' in df2['Event distance/length'].values[i]:
        arr1 = df2['Event distance/length'].values[i].split('k')
        df2['Event distance/length in km'].values[i] = arr1[0]
    elif 'mi' in df2['Event distance/length'].values[i]:
        arr2 = df2['Event distance/length'].values[i].split('mi')
        df2['Event distance/length in mi'].values[i] = arr2[0]
df2
```

```python
# converting 'Event distance/length in mi_' to float

df2['Event distance/length in mi_'] = df2['Event distance/length in mi'].astype(float)
df2
```

```python
# creating a new column for 'Event Country'

df2['Event Country'] = df.apply(lambda _: None, axis = 1)
```

```python
# separating the event country from the event name

for i in range(len(df2['Event name'])):
    if '(' in df2['Event name'].values[i]:
        arr_ = df2['Event name'].values[i].split('(')
        df2['Event Country'].values[i] = arr_[1]
        
#removing the closing brackets        
df2['Event Country'] = df2['Event Country'].str.rstrip(')')

# removing the bracket with the countries
df2['Event name'] = df2['Event name'].str[:-5].str.strip()

df2
```

```python
#dropping all rows with NaN values in the Athlete year of birth column
df3 = df2.dropna(subset=['Athlete year of birth'])
df3
```

```python
#converting data type to integer
df3['Athlete year of birth'].astype(int)
```

```python
#converting data type to integer
df3['Year of event'].astype(int)
```

```python
#checking the atheletes ages at the year the event took place

Athlete_age = []
for i in range(len(df3['Year of event'])):
    diff = df3['Year of event'].values[i] - df3['Athlete year of birth'].values[i]
    Athlete_age.append(diff)

df3['Athlete_age'] = Athlete_age
df3
```

```python
df3['Athlete age category'].unique()
```

```python
# dropping the 'Athlete age category' column

df4 = df3.dropna(subset=['Athlete age category'])
df4
```

```python
# stripping the strings in the 'Athlete age category' column

df4['Athlete age category'] = df4['Athlete age category'].str.lstrip('M')
```

```python
df4['Athlete age category'] = df4['Athlete age category'].str.lstrip('U')
```

```python
df4['Athlete age category'] = df4['Athlete age category'].str.lstrip('W')
```

```python
# converting the 'Athlete age category' to integer

df4['Athlete age category'].astype(int)
df4
```

```python
# checking the athlete's age and comparing them with the athlete's age category to see if they are qualified or not

qualification = []
a = 'qualified'
b = 'not_qualified'
for i in range(len(df4)):
    if int(df4['Athlete_age'].values[i]) > int(df4['Athlete age category'].values[i]):
        qualification.append(b)
    else:
        qualification.append(a)
df4['qualification'] = qualification
df4
```

```python
# replacing all None values in the 'Event distance/length in km' column with zero

for i in range(len(df4['Event distance/length in km'])):
    if df4['Event distance/length in km'].values[i] is None:
        df4['Event distance/length in km'].values[i] = '0'
```

```python
# removing all the string values in the 'Event distance/length in km' column and converting the numerials to float 


df4['Event distance/length in km'] = df4['Event distance/length in km'].replace('[^\d.]', '', regex = True).astype(float)
```

```python
# replacing all None values in the 'Event distance/length in mi' column with zero

for i in range(len(df4['Event distance/length in mi'])):
    if df4['Event distance/length in mi'].values[i] is None:
        df4['Event distance/length in mi'].values[i] = '0'
```

```python
# removing all the string values in the 'Event distance/length in mi' column and converting the numerials to float 

df4['Event distance/length in mi'] = df4['Event distance/length in mi'].replace('[^\d.]', '', regex = True).astype(float)
df4
```

```python
# converting miles to kilometre

mi_to_km = []
for i in range(len(df4['Event distance/length in mi'])):
    mul = df4['Event distance/length in mi'].values[i] * 1.6
    mi_to_km.append(mul)

df4['mi_to_km'] = mi_to_km
df4
```

```python
# merging the 'Event distance/length in km' column and the 'mi_to_km' column

All_in_km = []
for i in range(len(df4['Year of event'])):
    calc_ = df4['Event distance/length in km'].values[i] + df4['mi_to_km'].values[i]
    All_in_km.append(calc_)

df4['All_in_km'] = All_in_km
df4
```

```python
# dropping unnecessary columns

df4.drop(['Event distance/length', 'Event distance/length in km', 'Event distance/length in mi', 'Event distance/length in mi_', 'mi_to_km'], inplace = True, axis = 1)
df4
```

```python
# stripimg the string values in the 'Athlete performance' column

df4['Athlete performance'] = df4['Athlete performance'].str[:-2].str.strip()
```

```python
# creating new columns for hour, min, and sec

df4['hour'] = df.apply(lambda _: None, axis = 1)
df4['min'] = df.apply(lambda _: None, axis = 1)
df4['sec'] = df.apply(lambda _: None, axis = 1)
```

```python
# splitting the athlete's performance time into different columns 

for i in range(len(df4['Athlete performance'])):
    if ':' in df4['Athlete performance'].values[i]:
        arr_4 = df4['Athlete performance'].values[i].split(':')
        df4['hour'].values[i] = arr_4[0]
        df4['min'].values[i] = arr_4[1]
        df4['sec'].values[i] = arr_4[2]
df4
```

```python
# creating new columns for days and hours

df4['days'] = df4.apply(lambda _: None, axis = 1)
df4['hours'] = df4.apply(lambda _: None, axis = 1)
```

```python


for i in range(len(df4['hour'])):
    if 'd' in df4['hour'].values[i]:
        arr_5 = df4['hour'].values[i].split('d')
        df4['days'].values[i] = arr_5[0]
        df4['hour'].values[i] = arr_5[1]
df4
```

```python
#

df4.drop(['hours'], inplace = True , axis = 1)
```

```python
for i in range(len(df4['days'])):
    if df4['days'].values[i] is None:
        df4['days'].values[i] = '0'
df4
```

```python
df4['days'].astype(float)
df4['hour'].astype(float)
df4['min'].astype(float)
df4['sec'].astype(float)
df4
```

```python
mins_ = []
for i in range(len(df4['min'])):
    div = int(df4['min'].values[i]) / 60
    mins_.append(div)

df4['mins_'] = mins_
df4
```

```python
sec_ = []
for i in range(len(df4['sec'])):
    sss = int(df4['sec'].values[i]) / 3600
    sec_.append(sss)

df4['sec_'] = sec_
df4
```

```python
days_ = []
for i in range(len(df4['days'])):
    ddd = int(df4['days'].values[i]) * 24
    days_.append(ddd)

df4['days_'] = days_
df4
```

```python
hours_ = []
for i in range(len(df4['hour'])):
    hhh = int(df4['hour'].values[i]) * 1
    hours_.append(hhh)

df4['hours_'] = hours_
df4
```

```python
total_hours = []
for i in range(len(df4['hour'])):
    ttt = df4['hours_'].values[i] + df4['days_'].values[i] + df4['sec_'].values[i] + df4['mins_'].values[i]
    total_hours.append(ttt)

df4['total_hours'] = total_hours
df4
```

```python
df4.drop(['hour', 'min', 'sec', 'days', 'mins_', 'sec_', 'hours_', 'days_'], inplace = True, axis = 1)
df4

#hour	min	sec	days	mins_	sec_	days_	hours_m
```

```python
average_speed = []
for i in range(len(df4['All_in_km'])):
    asp = df4['All_in_km'].values[i] / df4['total_hours'].values[i]
    average_speed.append(asp)

df4['average_speed'] = average_speed
df4
```

```python
df4.drop(['Athlete average speed'], inplace = True, axis = 1)
df4
```

```python
# Print the rows with zero values

rows_with_zero = df4[df4['total_hours'] == 0]

rows_with_zero
```

```python
df4.drop(df4[df4['total_hours'] == 0].index, inplace=True)
df4
```

```python
mean_speed = df4.groupby(['Athlete country'])['average_speed'].mean().reset_index()

# Group by 'Athlete country' and calculate the count of appearances
appearances = df4.groupby(['Athlete country']).size().reset_index(name='Appearances')

# Merge the two DataFrames on 'Athlete country'
App_spd = pd.merge(mean_speed, appearances, on='Athlete country')

# Print the result
App_spd
```

```python
# Normalize average speed and appearances
App_spd['Normalized Speed'] = (App_spd['average_speed'] - App_spd['average_speed'].min()) / (App_spd['average_speed'].max() - App_spd['average_speed'].min())
App_spd['Normalized Appearances'] = (App_spd['Appearances'] - App_spd['Appearances'].min()) / (App_spd['Appearances'].max() - App_spd['Appearances'].min())

# Define weight factors
weight_speed = 0.5
weight_appearances = 0.5

# Calculate weighted score
App_spd['Weighted Score'] = weight_speed * App_spd['Normalized Speed'] + weight_appearances * App_spd['Normalized Appearances']

# Calculate performance score
App_spd['Performance Score'] = App_spd['Weighted Score']

# Sort by performance score
App_spd = App_spd.sort_values(by='Performance Score', ascending=False)

# Print the result
App_spd
```

```python
df_a_4 = App_spd.groupby('Athlete country')[['Performance Score']].mean()
top_10_countries__ = df_a_4.nlargest(10, 'Performance Score')
top_10_countries__
```

```python
top_10_countries__.to_csv('Country with the best performance.csv')
```

```python
# Visualisation
top_10_countries__.plot(kind = 'bar', color='deepskyblue')
plt.ylabel('Performance Score')
plt.xlabel('Athlete country')
plt.title('Best Performing countries')
plt.show
```
